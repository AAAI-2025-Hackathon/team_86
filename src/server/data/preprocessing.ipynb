```python
import pandas as pd
```

```python
UNIQUE_LABELS = ['Education', 'Travel', 'Sports', 'E-Commerce', 'Games', 'News', 'Health and Fitness', 'Computers and Technology', 'Food', 'Social Networking and Messaging']
```

```python
df = pd.read_csv('website_classification.csv')
df
```

```python
df = df.drop(columns={'Unnamed: 0', 'website_url'})
df = df.rename(columns={'Category': 'category', 'cleaned_website_text': 'keywords'})
df
```

```python
df = df[df['category'].isin(UNIQUE_LABELS)]
df
```

```python
df['category'].value_counts()
```

```python
has_nan = df.isna().any(axis=1)
pd.DataFrame(has_nan).assign(row_has_nan=True).loc[has_nan]
```

```python
from server.lib.constants import STOI

df = df.copy()
df['label'] = df['category'].apply(lambda x: STOI[x])
df = df.drop(columns={'category'})
```
